# Let's convert the data type of the "rate" column to float and remove the denominator.

```python
def handleRate(value):
    value=str(value).split('/')
    value=value[0];
    return float(value)

dataframe['rate'] = dataframe['rate'].apply(handleRate)
print(dataframe.head())
```
